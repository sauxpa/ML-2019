```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Stirling formula

$$
n! \sim \sqrt{2\pi n} \big(\frac{n}{e}\big)^n
$$

i.e

$$
\sum_{k=1}^n \log(k) \approx \frac{1}{2}\log(2\pi) + \frac{1}{2}\log(n) + n \log(\frac{n}{e})
$$

```python
n = int(1e3)

nn = np.linspace(1, n, n)
log_nn = np.log(nn)
log_factorial = np.cumsum(log_nn)
log_stirling = 0.5 * np.log(2 * np.pi) + 0.5 * np.log(nn) + nn * np.log(nn / np.exp(1))
log_lower_bound = nn * np.log(nn / np.exp(1))
log_upper_bound = 1 + np.log(nn) +  nn * np.log(nn / np.exp(1))
```

```python
with sns.axes_style('darkgrid'):
    fig, axes = plt.subplots(figsize=(7, 11), nrows=2, ncols=1)

    ax = axes[0]
    ax.semilogx(log_factorial, color='blue', label='log factorial')
    ax.semilogx(log_stirling, color='red', label='log stirling')
    ax.semilogx(log_lower_bound, color='orange', linestyle='--', label='log lower bound')
    ax.semilogx(log_upper_bound, color='green', linestyle='--', label='log upper bound')
    ax.legend(loc='upper left')
    
    ax = axes[1]
    ax.semilogx(log_factorial-log_stirling, color='blue', label='log error')
    ax.legend(loc='upper right')
    
    plt.tight_layout()
    plt.show()
```
